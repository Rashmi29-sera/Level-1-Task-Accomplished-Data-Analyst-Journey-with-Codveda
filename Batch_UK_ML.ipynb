```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('plant.csv')
```

```python
df
```

```python
plt.xlabel("Pesticide Amount")
plt.ylabel("Plant Production")
plt.scatter(df['Pesticide_Amount_mg_per_ha'],df['Plant_Production'], color='blue')
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
squaringFunction = PolynomialFeatures(degree=2)
Pesticide_square = squaringFunction.fit_transform(df[['Pesticide_Amount_mg_per_ha']])
```

```python
Pesticide_square
```

```python
parabola = linear_model.LinearRegression()
parabola.fit(Pesticide_square, df['Plant_Production'])
```

```python
plt.xlabel("Pesticide Amount")
plt.ylabel("Plant Production")
plt.scatter(df['Pesticide_Amount_mg_per_ha'],df['Plant_Production'], color='blue')
plt.plot(df['Pesticide_Amount_mg_per_ha'], parabola.predict(Pesticide_square), color = 'green')
```

# Unsupervised Learning

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    'Player': [
        'Joe Root', 'Ben Stokes', 'Jofra Archer', 'Jos Buttler', 'Chris Woakes',
        'Moeen Ali', 'James Anderson', 'Stuart Broad', 'Jonny Bairstow', 'Sam Curran',
        'Dawid Malan', 'Ollie Robinson', 'Mark Wood', 'Liam Livingstone', 'Adil Rashid'
    ],
    'Runs': [11000, 6000, 500, 4500, 3000, 3000, 650, 800, 5500, 1500, 2000, 300, 400, 1000, 1000],
    'Wickets': [50, 150, 90, 10, 150, 200, 680, 600, 5, 100, 2, 75, 120, 15, 250]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.scatter(df['Runs'], df['Wickets'], color='blue')
plt.xlabel('Runs')
plt.ylabel('Wickets')
```

```python
X = df[['Runs', 'Wickets']]
```

```python
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
df
```

```python
plt.scatter(df['Runs'], df['Wickets'], c=df['Cluster'])
plt.xlabel('Runs')
plt.ylabel('Wickets')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Runs', 'Wickets']])
```

```python
scaled_data
```

```python
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
centers = center = kmeans.cluster_centers_
centers = scaler.inverse_transform(centers)
```

```python
plt.scatter(df['Runs'], df['Wickets'], c=df['Cluster'], s=100, cmap='viridis')
# plt.scatter(centers[:,0], centers[:, 1], color='red', s=200, marker = 'X')
plt.xlabel('Runs')
plt.ylabel('Wickets')
plt.title('Cricketers Data Grouped')
plt.grid(True)
```

# Elbow Method

```python
distances = [] # Sum of distances for each number of cluster
```

```python
K = range(1,10) # 1, 2, 3, 4, 5, 6, 7
```

```python
for num in K:
  kmeans = KMeans(n_clusters=num)
  kmeans.fit(X)
  distances.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(K, distances, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Distances')
plt.xticks(K)
plt.grid('True')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
import pydicom
```

```python
dicom_path = '/content/Vida_Head.MR.Comp_DR-Gain_DR.1005.1.2021.04.27.14.20.13.818.14380335 (1).dcm'
```

```python
dataset = pydicom.dcmread(dicom_path)
```

```python
img_array = dataset.pixel_array.astype(float)
```

```python
img_array = (img_array - img_array.min()) / (np.max(img_array) - np.min(img_array))
```

```python
plt.figure(figsize=(10,5))
plt.imshow(img_array, cmap='gray')
plt.title('Original Image')
```

```python
h, w = img_array.shape
pixels = img_array.reshape(h*w, 1)
```

# Elbow Method

```python
distances = []
```

```python
K = range(2, 10) # 2,3,4,5,6,7,8,9
```

```python
for num in K:
  kmeans = KMeans(n_clusters=num)
  kmeans.fit(pixels)
  distances.append(kmeans.inertia_)
```

```python
plt.plot(K, distances, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Distances')
plt.xticks(K)
plt.grid('True')
```

```python
kmeans = KMeans(n_clusters=4)
labels = kmeans.fit_predict(pixels)
segmented_image = kmeans.cluster_centers_[labels].reshape(h, w)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(segmented_image, cmap='gray')
plt.title('Segmented Image')
```

```python
for i in range(4):
  cluster_filter = (labels.reshape(h,w)==i)
  segment = img_array * cluster_filter

  plt.figure()
  plt.imshow(segment, cmap='gray')
  #plt.title('Segment', i+1)
```
